Данная функция выгружает таблицу с кодами стран со сторонних сайтов.

```python
def country_codes_trad(req):
    #ищем коды стран в традиционном представлении
    #так как в процессе стало ясно, что сайт с данными использует свои обозначения, и следует приводить их к общeму виду
    soup = BeautifulSoup(req.text, 'lxml')
    html_table = soup.find_all("table")
    head, *table = soup.find_all("tr")
    columns = [th.text for th in head.find_all("th")]
    values = [[td.text for td in tr.find_all("td")] for tr in table]
    un_country_codes = pd.DataFrame(values, columns=columns)
    return un_country_codes
    
```

Здесь все остальные функции, которые нам пригодятся

```python
#Были сложности с парсингом сайта с данными о портах, поэтому пришлось обрабатывать исключения
def simple_parse_page(number_of_ports, ports_info):
    ports_list = dict()
    for i in range(0, 14 * number_of_ports, 14):
        port_info = []        
        for j in range(2, 13):
            res = ports_info[i + j].text
            if ports_info[i + j].text.isdigit():
                res =int(res)
            port_info.append(res) 
        port_info.append((driver.find_elements_by_css_selector(".ag-cell-content")[1]
                          .find_element_by_css_selector(".ag-cell-content-link").get_attribute("href")))      
        ports_list.update({ports_info[i + 1].text : port_info})
    return ports_list
    
def not_simple_parse_page(number_of_ports, ports_info):
    ports_list = dict()
    list_of_port_info = []
    if ports_info[2 * number_of_ports + 1].text.isdigit():
        k = 11
    else:
        k = 12
    for i in range(2 * number_of_ports, len(ports_info), k+1):
        port_info=[]
        for j in range(k):
            res = ports_info[i + j].text
            port_info.append(res)
        port_info.append((driver.find_elements_by_css_selector(".ag-cell-content")[1]
                          .find_element_by_css_selector(".ag-cell-content-link").get_attribute("href"))) 
        list_of_port_info.append(port_info)
    for i in range(0, 2 * number_of_ports, 2):
        ports_list.update({ports_info[i + 1].text: list_of_port_info[i // 2]})
    return ports_list

def parse_pages(pages_number):
    ports_list = dict()
    print(pages_number)
    for page in range(pages_number):
        print(f"Processing page {page}")
        time.sleep(4)
        ports_info = driver.find_elements_by_css_selector(".ag-cell-content")
        number_of_ports = len(ports_info)//14
        if ports_info[3].text.isdigit():
            ports_list.update(simple_parse_page(number_of_ports, ports_info))
        else:
            ports_list.update(not_simple_parse_page(number_of_ports, ports_info))
        button_right = driver.find_elements_by_css_selector(".MuiButton-sizeSmall")[3]
        ActionChains(driver).click(button_right).perform()
    return ports_list

def overall_ports_parse(i): #функция парсит страницы с портами всех стран и территорий
    un_country_codes = country_codes_inner()
    #переходим на сайт конкретной страны
    url_port_country = (url_port[:(url_port.find("|in|")+4)] + 
                        un_country_codes.iloc[i]["Country Name"].replace(" ", "%20") + 
                        url_port[(url_port.find("|in|")+4):url_port.find("flag_in=")+8]+ 
                        un_country_codes.iloc[i]["ISO 3166-1"] + 
                        url_port[url_port.find("flag_in=")+8:])
    driver.get(url_port_country)
    time.sleep(4)

    ports_list = dict()
    pages_number = driver.find_elements_by_css_selector(".MuiTypography-body1")
    pages_number = int(pages_number[9].text.split()[1])
    active_ports = int(driver.find_elements_by_css_selector(".MuiTypography-caption")[0].text.split()[1])
    if pages_number == 0:
        ports_list = {'-':['-']*12}
    else:
        ports_list = parse_pages(pages_number)
    if active_ports > 500:
        url_port_country = url_port_country.replace("desc", "asc")
        driver.get(url_port_country)
        ports_list.update(parse_pages(pages_number))
    
    ports = pd.DataFrame.from_dict(ports_list, orient='index')
    ports.reset_index(level=0, inplace=True) 
    ports.columns = ['Port','UN/LOCODE','Vessels In Port','Departures (last 24hrs)','Arrivals (last 24hrs)',
                     'Expected Arrivals','Local Time','Type','Related Anchorage',
                     'Area Global','Area Local','Also Known As', 'Link for the Details of the Port']

    url_port_country = (url_port[:(url_port.find("|in|")+4)] + 
                        un_country_codes.iloc[i]["Country Name"].replace(" ", "%20") + 
                        url_port[(url_port.find("|in|")+4):url_port.find("flag_in=")+8]+ 
                        un_country_codes.iloc[i]["ISO 3166-1"] + 
                        url_port[url_port.find("flag_in=")+8:])    
    url_port_country = url_port_country[:url_port_country.find("&vessels")]
    driver.get(url_port_country)
    time.sleep(4)
    all_ports = int(driver.find_elements_by_css_selector(".MuiTypography-caption")[0].text.split()[1])

    ports = ports.copy()
    ports.insert(0, "Overall Ports", all_ports)
    ports.insert(0, "ISO 3166-1", un_country_codes.iloc[i]["ISO 3166-1"])
    ports.insert(0, "Country", un_country_codes.iloc[i]["Country Name"])
    ports.sort_values("Vessels In Port", ascending=False, inplace = True)
    return ports

#Эта функция пригодится позже, когда будем считать неравномерность использования портов
def gini(df_country_ports):
    df_country_ports["Vessels In Port_cum"] = (np.cumsum(df_country_ports["Vessels In Port"] / 
                                                     df_country_ports["Overall Vessels In Country"]))
    df_country_ports["Ports_cum"] = (np.cumsum(df_country_ports["Port_counter"] / 
                                                         df_country_ports["Port_counter"].sum()))
    #Считаем табличным способом
    Ports_cum = df_country_ports['Ports_cum']
    Ports_cum_shifted = df_country_ports['Ports_cum'].shift().fillna(0)
    Vessels_cum = df_country_ports['Vessels In Port_cum']
    Vessels_cum_shifted = df_country_ports['Vessels In Port_cum'].shift().fillna(0)
    df_country_ports["Gini"] = 1-np.sum((Vessels_cum+Vessels_cum_shifted)*(Ports_cum - Ports_cum_shifted))
    return df_country_ports["Gini"]
```

Теперь данные приводятся к удобному виду. Также добавляются показатели, которые и будут визулизированы.

```python
df_ports = pd.read_json(r'C:\Users\asus\Desktop\Проект\Ports_DataFrame.json')

#Здесь находим территории, которые отображены на сайте, но которых нет в официальном источнике.
r = requests.get("https://www.unece.org/cefact/locode/service/location")
s = (pd.merge(country_codes_inner(), country_codes_trad(r), how='outer', on=['ISO 3166-1'])
                 .rename(columns={"Country Name_x": "Country Name (non-trad)", "Country Name_y": "Country Name"}))
exclusions = s.loc[s["Country Name"].isnull()]["Country Name (non-trad)"].to_dict()
#Уберем из датафрейма все записи с территориями, которых нет в официальном списке
for i in exclusions:
    df_ports = df_ports.loc[~(df_ports["Country"] == exclusions[i])]

s.drop(['Country Name (non-trad)'], axis='columns', inplace=True)
df_ports.drop(['Country'], axis='columns', inplace=True)  
df_ports = (pd.merge(country_codes_trad(r), df_ports, how='outer', on=['ISO 3166-1'])).rename(columns={"Country Name": "Country"})

#Приведем данные к единому типу
array =["Vessels In Port", "Departures (last 24hrs)", "Arrivals (last 24hrs)", "Expected Arrivals", "Port_counter"]
df_ports["Port_counter"] = (df_ports["Port"] != '-').astype(int)
for i in array:
    df_ports[i] = df_ports[i].astype(str)
    df_ports[i] = df_ports[i].str.replace("-", "0")
    df_ports[i] = df_ports[i].astype(int)
    
#Добавим столбцы с общим количеством портов и кораблей в данной стране
groupped = df_ports[["Country"] + array].groupby("Country").sum().reset_index()
df_ports= (pd.merge(df_ports, groupped, how='inner', on=['Country'], suffixes=("", "_x")))
df_ports.drop(['Port_counter', "Departures (last 24hrs)_x", 
               "Arrivals (last 24hrs)_x", "Expected Arrivals_x"], axis='columns', inplace=True)
df_ports = df_ports.rename(columns={"Port_counter_x": "Active Ports", "Vessels In Port_x":"Overall Vessels In Country"})

#Теперь найдем долю активных портов среди всех портов
df_ports["Share of Active Ports"] = 0
df_ports1 = df_ports[df_ports["Overall Ports"] == 0]
df_ports2 = df_ports[df_ports["Overall Ports"] > 0]
df_ports2["Share of Active Ports"] = df_ports2["Active Ports"] / df_ports2["Overall Ports"]
df_ports = pd.concat([df_ports1, df_ports2], axis=0, join='outer', ignore_index=True)
df_ports = df_ports.sort_values(by=['Country', "Overall Ports"])
df_ports.reset_index(inplace = True, drop=True)

#А теперь посчитаем коэффициент Джини неравномерности нагрузки портов. 
#В качестве метрики используем отношение судов в данном порте к количеству суден во всех портах страны
df_ports_gin = (df_ports[["Country",  "Port","Vessels In Port", "Overall Vessels In Country"]]
                .sort_values(by=["Country", "Vessels In Port"], ascending=True)
                .reset_index(drop=True))
df_ports_gin["Port_counter"] = 1
Gini_series = pd.Series()
for country in df_ports_gin["Country"].unique():
    df_country_ports = df_ports_gin[df_ports_gin["Country"] == country].reset_index(drop=True)
    Gini_series = pd.concat([Gini_series, gini(df_country_ports)])
Gini_series = Gini_series.reset_index(drop=True)
df_ports["Gini"] = Gini_series

#Здесь снова стало понятно, что надо изменить код стран.(Alpha-3 code 2 знака и 3 знака)
#ищем коды стран в традиционном представлении
#так как в процессе стало ясно, что сайт с данными использует свои обозначения и надо парсить изнутри
r = requests.get("https://www.iban.com/country-codes")
codes = country_codes_trad(r).drop(["Country", "Numeric"], axis=1)
df_ports = pd.merge(codes, df_ports, how='outer', left_on=['Alpha-2 code'], right_on=["ISO 3166-1"])
df_ports.drop(df_ports.tail(1).index,inplace=True)
df_ports.drop(columns=["Alpha-2 code"],inplace=True)
df_ports.drop(df_ports[df_ports["Alpha-3 code"] == "BVT"].index[0],inplace=True)
df_ports = (df_ports.replace('United States\xa0\xa0\xa0\xa0[A to E]\xa0\xa0\xa0\xa0[F to J]\xa0\xa0\xa0\xa0[K to O]\xa0\xa0\xa0\xa0[P to T]\xa0\xa0\xa0\xa0[U to Z]'
                  , "United States"))
df_ports
```

Наконец, посмотрим, на соотношение географических параметров территорий и данных о портах. Далее собираем данные и переходим в R.

```python
url = "https://en.wikipedia.org/wiki/List_of_countries_by_length_of_coastline"
r = requests.get(url)
soup = BeautifulSoup(r.text, "lxml")
table = soup.find_all("table")[0]
df = pd.read_html(str(table), header = 0)[0].rename(columns={"The World Factbook[3].1":'The World Factbook', 
                                                            "World Resources Institute[2].1": "World Resources Institute",
                                                            "Land area km2 (TWF)[4]":"Land area km2 (TWF)"})
df.drop(['The World Factbook[3]', 'World Resources Institute[2]', "Coast/area ratio (m/km2).1"], 
        axis='columns', inplace=True)
df = df[3:].fillna('0')
df.reset_index(inplace = True, drop =True)
for col in df.columns:
    for i in range(len(df[col].index)):
        country = df[col].loc[i]
        if country.find('[') != -1:
            df[col].loc[i] = country[:country.find('[')]
        if df[col].loc[i] == '':
            df[col].loc[i] = '0'
for i in df.columns[1:]:
    df[i] = df[i].str.replace(",", "").str.replace(" ", "")
    df[i] = df[i].astype('float')
df.set_index("Country")
df["Length of Coast line"] = df[["The World Factbook", "World Resources Institute"]].max(axis=1)
df["Coast/area ratio (m/km2)"] = df["Length of Coast line"] * 1000 / df["Land area km2 (TWF)"]
df
```
